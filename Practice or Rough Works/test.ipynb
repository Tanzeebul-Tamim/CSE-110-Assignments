```python
while True:
    sequence = []
    element = ""
    user_input = input(
        "Please enter a number sequence separated by space, or 'STOP' to end: "
    )
    if user_input.upper() == "STOP":
        break

    for i in range(len(user_input)):
        char = user_input[i]
        if ("0" <= char <= "9" or char == "-") and char != chr(32):
            element += char
        elif char == chr(32):
            sequence.append(int(element))
            element = ""
        else:
            raise ValueError("The sequence must contain numbers only")

    if not chr(32) in user_input:
        raise ValueError("The sequence must contain more than one number")

    sequence.append(int(element))

    abs_sequence = []
    ref_seq = []
    stored_for_abs = 0
    for j in range(len(sequence)):
        char = sequence[j]
        if j == 0:
            stored_for_abs = char
        else:
            ref_seq.append(j)
            abs_diff = stored_for_abs - char
            if abs_diff < 0:
                abs_diff *= -1
            abs_sequence.append(abs_diff)
            stored_for_abs = char

    def check_UB_Jumper():
        for k in range(len(abs_sequence)):
            found = False
            for l in range(len(ref_seq)):
                if abs_sequence[k] == ref_seq[l]:
                    found = True
                    break
            if not found:
                return False
        return True

    result = check_UB_Jumper()
    if result:
        print("UB Jumper")
    else:
        print("Not UB Jumper")
```

```python
def sort_alpha_numerics(alpha_num):
    lowers = []
    uppers = []
    even_digits = []
    odd_digits = []

    for i in range(len(alpha_num)):
        char = alpha_num[i]
        ascii_value = ord(char)
        if 97 <= ascii_value <= 122:
            lowers.append(char)
        elif 65 <= ascii_value <= 90:
            uppers.append(char)
        elif 48 <= ascii_value <= 57:
            if ascii_value % 2 == 0:
                even_digits.append(char)
            else:
                odd_digits.append(char)


    def sort_characters(list):
        for j in range(len(list)):
            min_index = j
            for k in range(j + 1, len(list)):
                if ord(list[k]) < ord(list[min_index]):
                    min_index = k
            temp = list[j]
            list[j] = list[min_index]
            list[min_index] = temp

        result_string = ""
        for l in range(len(list)):
            result_string += list[l]

        return result_string


    sorted_lowers = sort_characters(lowers)
    sorted_uppers = sort_characters(uppers)
    sorted_odd_digits = sort_characters(odd_digits)
    sorted_even_digits = sort_characters(even_digits)
    result = sorted_lowers + sorted_uppers + sorted_odd_digits + sorted_even_digits
    
    return result

result = sort_alpha_numerics("Bracu1234")
print(result)
```
